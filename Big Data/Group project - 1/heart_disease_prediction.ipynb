# Data Cleanup

```python
print(heart_data['ca'].value_counts())
print(heart_data['thal'].value_counts())
```

#### We dont have Null values, but we have '?' value present in two fields. We are using Mode to fill the '?'.

```python
heart_data['ca'] = heart_data['ca'].replace('\?',np.NaN, regex=True)
heart_data['thal'] = heart_data['thal'].replace('\?',np.NaN, regex=True)
```

```python
print(heart_data['ca'].value_counts())
print(heart_data['thal'].value_counts())
```

```python
heart_data.isnull().sum()
```

```python
heart_data['ca'].fillna(heart_data['ca'].mode()[0],inplace=True)
heart_data['thal'].fillna(heart_data['thal'].mode()[0],inplace=True)
```

```python
print(heart_data['ca'].value_counts())
print(heart_data['thal'].value_counts())
```

```python
heart_data.isnull().sum()
```

#### Missing value has been taken care of. We have to update the target variable to just 0 and 1
#### 1,2,3 and 4 means Heart disease. So we are updating all 2,3 and 4 to 1.

```python
heart_data['num'].value_counts()
```

```python
heart_data['num'] = heart_data['num'].replace(2,1, regex=True)
heart_data['num'] = heart_data['num'].replace(3,1, regex=True)
heart_data['num'] = heart_data['num'].replace(4,1, regex=True)
```

```python
heart_data['num'].value_counts()
```

# Feature Engineering

#### Assign the values to X and Y set for model to take as an input

```python
X = heart_data.iloc[:,:-1].values
Y = heart_data.iloc[:,-1].values
```

#### Split the Test and Train with 80:20 ratio

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=0)
```

# Model Building

#### We have the classification problem with the target value. we need used the classification alogorithm.

#### 1. Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
model = LogisticRegression()
model.fit(x_train,y_train)
```

#### Feeding the Test data into our model to predict the output

```python
test_prediction = model.predict(x_test)
```

#### Display the predicted output of test data

```python
test_prediction
```

#### Display the actual output values of test data

```python
y_test
```

#### Compare the actual output and predicted output and find the accurary of our model.

```python
accuracy_score(y_test,test_prediction)
```

#### 2. Naive Bayes Alogorithm

```python
from sklearn.naive_bayes import GaussianNB
nb_model = GaussianNB()
nb_model.fit(x_train, y_train)
```

#### Feed the test data to our model to predict the output value

```python
naive_pred = nb_model.predict(x_test)
```

#### Compre the actual value and the predicted value to find the accuracy

```python
accuracy_score(y_test,naive_pred)
```

### 3 . Support Vector Machine model

```python
from sklearn.svm import SVC
svc_model = SVC(kernel = 'rbf')
svc_model.fit(x_train, y_train)
```

#### Feed the test data to our model to predict the output value

```python
svc_pred = svc_model.predict(x_test)
```

#### Compre the actual value and the predicted value to find the accuracy

```python
accuracy_score(y_test,svc_pred)
```

#### From the three classsification algorithm we used, we have find out that Logistic regression gives the better accuracy.
